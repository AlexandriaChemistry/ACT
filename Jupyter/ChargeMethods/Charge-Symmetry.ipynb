Charge generation with symmetry
===========================

Often we want to have the same charges on symmetry related atoms, for instance the H in a methyl group. This can likely be accomplished using [Lagrange multipliers](https://en.wikipedia.org/wiki/Lagrange_multiplier) in the equation. For the case of methanol, atoms 2, 3 and 4 should have the same charge. We then write:

$E_{SQE} = E_{EEM} + \sum\limits_{i,j}^{M} \left(\frac{1}{2}\zeta_{ij}p_{ij}^2+\Delta\chi_{ij}(q_i-q_j)\right) + \lambda_1(q_2-q_3) + \lambda_2(q_2-q_4)$

Since $q_2$ is equal to $q_3$ and $q_4$, we are not adding anything to the energy.  
First, let's make the equation general, there are $N_c$ constraints on the charges:

$E_{SQE} = E_{EEM} + \sum\limits_{i,j}^{M} \left(\frac{1}{2}\zeta_{ij}p_{ij}^2+\Delta\chi_{ij}(q_i-q_j)\right) + \sum\limits_{c=1}^{N_c}\lambda_c\left(q_{c_1}-q_{c_2}\right)$

we then have to insert the equations for the charges, and get

$E_{SQE} = \sum\limits_{n=1}^{N} \left[\left(\frac{q_{tot}}{N} + \sum\limits_{m=1}^{M_n} p_{nm}\right)\left(\chi_n + \frac{1}{2} \eta_n \left(\frac{q_{tot}}{N} + \sum\limits_{m=1}^{M_n} p_{nm}\right) + \frac{1}{2}\sum\limits_{l\ne n}^{N}\left[\frac{q_{tot}}{N} + \sum\limits_{m=1}^{M_l} p_{lm}\right]J_{nl}\right)\right]+\sum\limits_{i,j}^{M}\left[\frac{1}{2}\zeta_{ij}p_{ij}^2+\Delta\chi_{ij}\left(\left[\frac{q_{tot}}{N} + \sum\limits_{m=1}^{M_i} p_{im}\right]-\left[\frac{q_{tot}}{N} + \sum\limits_{m=1}^{M_j} p_{jm}\right]\right)\right] + \sum\limits_{c=1}^{N_c}\lambda_c\left(\sum\limits_{m=1}^{M_{c_1}} p_{jc_1} - \sum\limits_{n=1}^{M_{c_2}} p_{c_2n}\right)$
Here, for the first constraint we have $c_1$ = 2, $c_2$ = 3
and for the second one $c_1$ = 2, $c_2$ = 4.

Now we have to take the derivatives with respect to $p_{ij}$ as well as to $\lambda_c$ and solve the set of equations.
